```python
import os
import pyspark

# Autodetekcja wersji Sparka → właściwy connector
spark_version = pyspark.__version__
print(f"Wykryta wersja PySpark: {spark_version}")

if spark_version.startswith("4"):
    KAFKA_PACKAGE = "org.apache.spark:spark-sql-kafka-0-10_2.13:4.0.0-preview2"
else:
    # Spark 3.x
    KAFKA_PACKAGE = "org.apache.spark:spark-sql-kafka-0-10_2.12:3.5.0"

print(f"Użyty connector:        {KAFKA_PACKAGE}")

# Musi być ustawione PRZED SparkSession.builder
os.environ['PYSPARK_SUBMIT_ARGS'] = f'--packages {KAFKA_PACKAGE} pyspark-shell'
```

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Lab4-Kafka")
    .config("spark.jars.packages", KAFKA_PACKAGE)
    .getOrCreate()
)

spark.sparkContext.setLogLevel("WARN")

print(f"Spark {spark.version} — gotowy")
```

```python
kafka_raw = (
    spark.readStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "broker:9092")
    .option("subscribe", "transactions")
    .load()
)

kafka_raw.printSchema()
```

```python
def process_batch(df, batch_id, tstop=5):
    print(f"Batch ID: {batch_id}")
    df.show(truncate=False)

kafka_raw = (
    spark.readStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "broker:9092")
    .option("subscribe", "transactions")
    .load()
)

query = (
    kafka_raw.writeStream
    .format("console")
    .outputMode("append")
    .foreachBatch(process_batch)
    .option("truncate", False)
    .start()
)
```

```python
query.stop()
```

```python
from pyspark.sql.functions import col

df = kafka_raw.select(
    col("timestamp").alias("time"),
    col("value").cast("string")
)

query = (
    df.writeStream
    .format("console")
    .outputMode("append")
    .foreachBatch(process_batch)
    .option("truncate", False)
    .start()
)
```

```python
query.stop()
```

```python
from pyspark.sql.types import StructType, StructField, StringType, DoubleType
from pyspark.sql.functions import from_json, col

tx_schema = StructType([
    StructField("tx_id",     StringType()),
    StructField("user_id",   StringType()),
    StructField("amount",    DoubleType()),
    StructField("store",     StringType()),
    StructField("category",  StringType()),
    StructField("timestamp", StringType()),
])

step2 = kafka_raw.select(
    from_json(col("value").cast("string"), tx_schema).alias("tx")
)

query = (
    step2.writeStream
    .format("console")
    .outputMode("append")
    .foreachBatch(process_batch)
    .option("truncate", False)
    .start()
)
```

```python
query.stop()
```

```python
from pyspark.sql.functions import to_timestamp

df = (
    kafka_raw
    .select(from_json(col("value").cast("string"), tx_schema).alias("tx"))
    .select("tx.*")
    .withColumn("timestamp", to_timestamp("timestamp"))
)

print("Finalny schemat:")
df.printSchema()
```

```python
from pyspark.sql.functions import window, count, sum as _sum, round as _round

windowed = (
    df
    .withWatermark("timestamp", "3 seconds")
    .groupBy(window("timestamp", "20 seconds"), "store")
    .agg(
        count("tx_id").alias("liczba_tx"),
        _round(_sum("amount"), 2).alias("suma_PLN"),
    )
)
```

```python
query = (
    windowed.writeStream
    .format("console")
    .outputMode("append")
    .option("truncate", False)
    .start()
)
```

```python
query.stop()
```

```python
from pyspark.sql.functions import to_json, struct, lit

alerts = (
    df
    .filter(col("amount") > 3000)
    .select(
        to_json(
            struct(
                "tx_id", "user_id", "amount", "store", "category",
                col("timestamp").cast("string"),
                lit("HIGH").alias("alert_level"),
            )
        ).alias("value")
    )
)

alert_query = (
    alerts.writeStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "broker:9092")
    .option("topic", "alerts")
    .outputMode("append")
    .start()
)

print("Strumień alertów uruchomiony.")
```

```python
from pyspark.sql.functions import to_json, struct, lit

alerts = (
    df
    .filter(col("amount") > 3000)
    .select(
        to_json(
            struct(
                "tx_id", "user_id", "amount", "store", "category",
                col("timestamp").cast("string"),
                lit("HIGH").alias("alert_level"),
            )
        ).alias("value")
    )
)

alert_query = (
    alerts.writeStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "broker:9092")
    .option("topic", "alerts")

    .option("checkpointLocation", "/tmp/checkpoints_alerts")

    .outputMode("append")
    .start()
)

print("Strumień alertów uruchomiony.")
```

```python
# 1. Dlaczego w trybie 'append' pierwsze wyniki pojawiają się z opóźnieniem?
# append pokazuje wyniki dopiero po zamknięciu okna.
# Spark czeka na watermark i ewentualne spóźnione dane.

# 2. Co się stanie jeśli ustawisz watermark na "0 seconds"?
# Spark nie będzie czekał na spóźnione dane.
# Spóźnione rekordy zostaną odrzucone natychmiast.

# 3. Jaka jest różnica między window() w Lab 2 (batch) a window() tutaj (streaming)?
# W batch operujemy na skończonym zbiorze danych.
# W streamingu dane napływają cały czas i okna są aktualizowane dynamicznie.

# 4. Dlaczego do zapisu do Kafki używamy to_json(), a nie po prostu select("value")?
# Kafka przyjmuje dane jako string/bytes.
# to_json() zamienia struct Spark na JSON string.
```

```python
import os
import pyspark

spark_version = pyspark.__version__

if spark_version.startswith("4"):
    KAFKA_PACKAGE = "org.apache.spark:spark-sql-kafka-0-10_2.13:4.0.0-preview2"
else:
    KAFKA_PACKAGE = "org.apache.spark:spark-sql-kafka-0-10_2.12:3.5.0"

os.environ["PYSPARK_SUBMIT_ARGS"] = f"--packages {KAFKA_PACKAGE} pyspark-shell"
```

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Lab4-Kafka")
    .getOrCreate()
)

spark.sparkContext.setLogLevel("WARN")
```

```python
kafka_raw = (
    spark.readStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "broker:9092")
    .option("subscribe", "transactions")
    .load()
)

kafka_raw.printSchema()
```

```python
from pyspark.sql.functions import col

df_string = kafka_raw.select(
    col("value").cast("string").alias("json")
)

df_string.printSchema()
```

```python
from pyspark.sql.functions import from_json, col
from pyspark.sql.types import StructType, StructField, StringType, DoubleType

tx_schema = StructType([
    StructField("tx_id", StringType()),
    StructField("user_id", StringType()),
    StructField("amount", DoubleType()),
    StructField("store", StringType()),
    StructField("category", StringType()),
    StructField("timestamp", StringType())
])

df_parsed = df_string.select(
    from_json(col("json"), tx_schema).alias("tx")
).select("tx.*")
```

```python
df_parsed.writeStream \
    .format("console") \
    .outputMode("append") \
    .start()
```

```python
query = (
    df_parsed.writeStream
    .foreachBatch(lambda df, id: df.show(truncate=False))
    .outputMode("append")
    .start()
)

query.awaitTermination()
```

```python
query.stop()
```

```python
from pyspark.sql.functions import window, count, sum as _sum

windowed = (
    df_parsed
    .withWatermark("timestamp", "3 seconds")
    .groupBy(
        window("timestamp", "2 minutes", "1 minute"),
        "store"
    )
    .agg(
        count("tx_id").alias("cnt"),
        _sum("amount").alias("sum")
    )
)
```

```python
query2 = (
    windowed.writeStream
    .format("console")
    .outputMode("append")
    .start()
)

query2.awaitTermination()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *

KAFKA_PACKAGE = "org.apache.spark:spark-sql-kafka-0-10_2.13:4.0.0-preview2"

spark = (
    SparkSession.builder
    .appName("Lab4-Kafka")
    .config("spark.jars.packages", KAFKA_PACKAGE)
    .getOrCreate()
)

spark.sparkContext.setLogLevel("WARN")
```

```python
kafka_raw = (
    spark.readStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "broker:9092")
    .option("subscribe", "transactions")
    .load()
)
```

```python
tx_schema = StructType([
    StructField("tx_id", StringType()),
    StructField("user_id", StringType()),
    StructField("amount", DoubleType()),
    StructField("store", StringType()),
    StructField("category", StringType()),
    StructField("timestamp", StringType()),
])
```

```python
df = (
    kafka_raw
    .select(from_json(col("value").cast("string"), tx_schema).alias("tx"))
    .select("tx.*")
    .withColumn("timestamp", to_timestamp("timestamp"))
)
```

```python
from pyspark.sql.functions import to_json, struct, lit, col

alerts = (
    df
    .filter(col("amount") > 3000)
    .select(
        to_json(
            struct(
                "tx_id",
                "user_id",
                "amount",
                "store",
                "category",
                col("timestamp").cast("string"),
                (col("amount") / 400.0).alias("ratio"),
                lit("HIGH").alias("alert_level")
            )
        ).alias("value")
    )
)
```

```python
alert_query = (
    alerts.writeStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "broker:9092")
    .option("topic", "alerts")
    .option("checkpointLocation", "/tmp/checkpoints_alerts")
    .outputMode("append")
    .start()
)

print("Strumień alertów uruchomiony")
```

```python
alerts = (
    df
    .filter(col("amount") > 3000)
    .select(
        to_json(
            struct(
                "tx_id",
                "user_id",
                "amount",
                (col("amount") / 400.0).alias("ratio"),
                "store",
                "category",
                col("timestamp").cast("string"),
                lit("HIGH").alias("alert_level"),
            )
        ).alias("value")
    )
)
```

```python
alert_query = (
    alerts
    .writeStream
    .format("kafka")
    .option("kafka.bootstrap.servers", "localhost:9092")
    .option("topic", "alerts")
    .option("checkpointLocation", "/tmp/check_alerts")
    .start()
)
```

```python
# Po zatrzymaniu producenta nowe dane przestają napływać do streamu.
# Przez chwilę Spark nadal pokazuje wyniki, ponieważ dane są jeszcze
# obecne w oknie czasowym 2 minut.
# Po około 2 minutach stare rekordy wypadają z okna,
# więc agregacje i alerty przestają się pojawiać.
# Dzieje się tak dlatego, że sliding window operuje tylko
# na danych znajdujących się aktualnie w zakresie okna czasowego.
```
